```python
import wrds
import pandas as pd

db = wrds.Connection()
```

```python
tesla = db.raw_sql("""
SELECT 
    gvkey,
    conm,
    datadate,
    fyear,
    sale,
    at,
    ni,
    ib,
    oibdp,
    lt,
    ceq,
    xrd,
    capx,
    csho,
    prcc_f
FROM comp.funda
WHERE gvkey = '184996'
  AND indfmt = 'INDL'
  AND datafmt = 'STD'
  AND popsrc = 'D'
  AND consol = 'C'
  AND fyear BETWEEN 2022 AND 2024
ORDER BY fyear
""")

tesla
```

```python
toyota = db.raw_sql("""
SELECT 
    gvkey,
    conm,
    datadate,
    fyear,
    sale,
    at,
    ni,
    ib,
    oibdp,
    lt,
    ceq,
    xrd,
    capx,
    csho,
    prcc_f
FROM comp.funda
WHERE gvkey = '019661'
  AND fyear > 2021
ORDER BY fyear
""")

toyota
```

```python
ford = db.raw_sql("""
SELECT 
    gvkey,
    conm,
    datadate,
    fyear,
    sale,
    at,
    ni,
    ib,
    oibdp,
    lt,
    ceq,
    xrd,
    capx,
    csho,
    prcc_f
FROM comp.funda
WHERE gvkey = '004839'
  AND indfmt = 'INDL'
  AND datafmt = 'STD'
  AND consol = 'C'
  AND popsrc = 'D'
  AND fyear BETWEEN 2022 AND 2024
ORDER BY fyear
""")

ford
```

```python
BYD = db.raw_sql("""
SELECT 
    gvkey,
    conm,
    datadate,
    fyear,
    sale,
    at,
    ib,
    oibdp,
    lt,
    ceq,
    xrd,
    g_funda.cshr
    capx,
    g_funda.prc
FROM comp_global_daily.g_funda
WHERE gvkey = '253759'
  AND fyear between 2022 and 2024
ORDER BY fyear
""")

BYD
```

```python
BYD['company'] = 'BYD'
Tesla['company'] = 'Tesla'
Toyota['company'] = 'Toyota'
Ford['company'] = 'Ford'

df = pd.concat([BYD, Tesla, Toyota, Ford], ignore_index=True)
df = df.rename(columns={
    'conm': 'company_name',
    'datadate': 'date',
    'fyear': 'year',
    'sale': 'revenue',
    'at': 'total_assets',
    'ib': 'net_income',
    'oibdp': 'operating_income',
    'lt': 'total_liabilities',
    'ceq': 'equity',
    'xrd': 'r_and_d',
    'cshr': 'shares_outstanding',
    'capx': 'capital_expenditure',
    'prc': 'share_price'
})
print(df.columns)
print(df.shape)
display(df.head())
display(df[['company', 'year']].drop_duplicates().sort_values(['company', 'year']))
print(df.isnull().sum())
```

```python
df = df.sort_values(['company', 'year'])

df['net_profit_margin'] = df['net_income'] / df['revenue']
df['roa'] = df['net_income'] / df['total_assets']
df['roe'] = df['net_income'] / df['equity']
df['debt_ratio'] = df['total_liabilities'] / df['total_assets']
df['debt_to_equity'] = df['total_liabilities'] / df['equity']
df['r_and_d_ratio'] = df['r_and_d'] / df['revenue']
df['capex_ratio'] = df['capital_expenditure'] / df['revenue']

df['revenue_growth'] = df.groupby('company')['revenue'].pct_change()
df['net_income_growth'] = df.groupby('company')['net_income'].pct_change()
```
